```python
from qutip import *
import numpy as np
```

```python
def to_cart(angle):
    theta, phi = angle
    x = np.sin(theta) * np.cos(phi) 
    y = np.sin(theta) * np.sin(phi) 
    z = np.cos(theta) 
    return np.array([x, y, z])
```

```python
b = Bloch()
b.sphere_alpha = 0
b.frame_alpha = 0.05
b.xlabel = ['', '']
b.ylabel = ['', '']
b.figsize = [5, 5]
theta = np.pi/6
phi = np.pi/3

angles = [[theta, phi], 
          [theta + np.pi/2, phi],
          [np.pi / 2, phi + np.pi / 2]]
vectors = [to_cart(angle) for angle in angles]
b.add_vectors(vectors)
b.add_annotation(vectors[0], 'Z', verticalalignment='bottom', color='#58508d')
b.add_annotation(vectors[1], 'X', verticalalignment='top', color='#ff6361')
b.add_annotation(vectors[2], 'Y', horizontalalignment='left', color='#bc5090')


b.vector_color = ['#58508d', '#ff6361', '#bc5090']
b.show()
```
